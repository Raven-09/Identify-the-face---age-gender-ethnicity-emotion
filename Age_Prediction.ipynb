```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_frame = pd.read_csv('final_dataset.csv')
data_frame.head(10)
```

```python
data_frame['Age'].value_counts()
```

```python
data_frame = data_frame[data_frame['Age']!="Emotion_Neutral"]
data_frame['Age'].value_counts()
```

```python
data_frame = data_frame[data_frame['Age']!="Emotion_Happy"]
data_frame['Age'].value_counts()
```

```python
data_frame = data_frame[data_frame['Age']!="0"]
data_frame['Age'].value_counts()
```

```python
X = data_frame.iloc[:,1:2501]
y = data_frame['Age']
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn import preprocessing
lb = preprocessing.LabelBinarizer()
y_train = lb.fit_transform(y_train)
y_test = lb.transform(y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
y_train
```

```python
from keras.layers import Dense
from keras.layers import MaxPooling2D
from keras.layers import Convolution2D
from keras.layers import Flatten
from keras.models import Sequential
```

```python
model = Sequential()
model.add(Convolution2D(30,3, input_shape = (50,50,1), activation = 'relu', padding = 'same'))
model.add(MaxPooling2D(pool_size= (2,2)))
model.add(Convolution2D(40,3, activation = 'relu', padding = 'same'))
model.add(Convolution2D(45,3, activation = 'relu', padding = 'same'))
model.add(MaxPooling2D(pool_size= (2,2))) 
model.add(Convolution2D(50,3, activation = 'relu', padding = 'same'))
model.add(Flatten())
model.add(Dense(output_dim = 128, activation = 'relu'))
model.add(Dense(output_dim = 5, activation = 'softmax'))
```

```python
model.compile(optimizer='adam', loss = 'categorical_crossentropy', metrics  = ['accuracy'])
```

```python
trainX = X_train.reshape(X_train.shape[0],50, 50, 1).astype( 'float32' )
testX = X_test.reshape(X_test.shape[0],50, 50, 1).astype( 'float32' )
```

```python
model.fit(trainX, y_train, batch_size=50, epochs = 10)
```

```python
score = model.evaluate(testX, y_test, batch_size=10)
```

```python
score
```

```python
model_json = model.to_json()
with open("age.json", "w") as json_file:
    json_file.write(model_json)
    
#serialise weights as HDF5

model.save_weights("age.h5")
print("Saved")
```
