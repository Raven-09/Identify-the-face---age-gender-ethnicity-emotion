```python
import numpy as np
import pandas as pd
import json
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('dataset.json') as f:
    data = json.load(f)
```

```python
import requests
```

```python
ls = []

for i in range(0, 2500):
    ls.append(i)
```

```python
ls.append('Emotion')
ls.append('Age')
ls.append('Complexion')
ls.append('Gender')
```

```python
len(ls)
```

```python
df = pd.DataFrame(columns=ls)
df
```

```python
from PIL import Image
import PIL

for i in range(len(data)):
    img_data = requests.get(data[i]['content']).content
    print(i)

    with open('pic1.jpg', 'wb') as ff:
         ff.write(img_data)
    
    img = Image.open('pic1.jpg').convert('L')
    arr = np.array(img)

    
    for j in range(len(data[i]['annotation'])):
        
        imageHeight = data[i]['annotation'][0]['imageHeight']
        imageWidth = data[i]['annotation'][0]['imageWidth']
    
        x1 = int(data[i]['annotation'][j]['points'][0]['x'] * imageWidth)
        y1 = int(data[i]['annotation'][j]['points'][0]['y'] * imageHeight)
        x2 = int(data[i]['annotation'][j]['points'][1]['x'] * imageWidth)
        y2 = int(data[i]['annotation'][j]['points'][1]['y'] * imageHeight)

        arr1 = arr[y1:y2, x1:x2]

        img2 = Image.fromarray(arr1)

        img2 = img2.resize((50,50), PIL.Image.ANTIALIAS)

        arr1 = np.array(img2)
        
        face = []
        for k in range(50):
            for _k in range(50):
                face.append(arr1[k][_k])
                
        for k in data[i]['annotation'][j]['label']:
            face.append(k)
        
        while len(face) < 2504:
            face.append(0)
            
        face = tuple(face)
        ls1 = tuple(ls)
        
        #temp_data = pd.DataFrame()
        
        df = df.append(dict(zip(ls1, face)), ignore_index = True)
```
